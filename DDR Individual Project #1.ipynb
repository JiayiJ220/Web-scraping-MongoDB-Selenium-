```python
## 1.2 ##



def saveString(html, filename):
    try:
        file = open(filename,"w")
        file.write(str(html))
        file.close()
    except Exception as ex:
        print('Error: ' + str(ex))


        
def loadString(filename):
    try:
        html = open(filename, "r", encoding='utf-8').read()
        return(html)
    except Exception as ex:
        print('Error: ' + str(ex))        
        
        
    
      
def main():
    try:
        
        ##################################  1.2 Part (a) ##################################
      
        url_01 = "https://www.ebay.com/sch/i.html?_from=R40&_nkw=amazon+gift+card&LH_Sold=1"
        header = {'User-Agent':'Mozilla/5.0'}
        time.sleep(10)
        page_01 = requests.get(url_01, headers = header)
        
        # Create a beautifulsoup object 
        doc_01 = BeautifulSoup(page_01.content, "html.parser")
        filename_01 = "amazon_gift_card_01.htm"
        saveString(doc_01, filename_01)
        time.sleep(10)
        
        ##################################  1.2 Part (b) ##################################
        
        for i in range (0,10):
            index = i+1
            url_pages = "https://www.ebay.com/sch/i.html?_from=R40&_nkw=amazon+gift+card&LH_Sold=1&_pgn="+ str(index)
            header = {'User-Agent':'Mozilla/5.0'}
            page_loop = requests.get(url_pages, headers = header)
            
            # Create a beautifulsoup object 
            doc_pages = BeautifulSoup(page_loop.content, "html.parser")
            
            if index <= 9:
                filename_loop = "amazon_gift_card_0" + str(index) + ".htm"
            else:
                filename_loop = "amazon_gift_card_" + str(index) + ".htm"
            
            saveString(doc_pages, filename_loop)
            
            time.sleep(10)
            
            
        ##################################  1.2 Part (c) ##################################
        
        for g in range(0,10):
            fileindex1 = g+1
            if fileindex1 <= 9:
                filename_partc = "amazon_gift_card_0" + str(fileindex1) + ".htm"
            else:
                filename_partc = "amazon_gift_card_" + str(fileindex1) + ".htm"
            
            files_partc = loadString(filename_partc)
            soup_page_pc = BeautifulSoup(files_partc, "html.parser")
        
        
        ##################################  1.2 Part (d) ##################################
        
            gc_item = soup_page_pc.select("div.s-item__wrapper.clearfix")
            item_list = gc_item[1:]
            for j in range(0,60):
                item = item_list[j]
                
                ## Find the title:

                name = item.select("div.s-item__title")
                item_title = []
                for i in name:
                    item_title.append(i.text)
                item_title_string = str(item_title)
                gf_title = re.sub("New Listing", "", item_title_string)
                
    
    
    
                ## Find the price:
    
                price = item.select("span.s-item__price")
                item_price = []
                for i in price:
                    item_price.append(i.text)
             
    
    
    
                ## Find the shipping price: 
                item_test = str(item)
                if "s-item__shipping s-item__logisticsCost" in item_test:
                    shipping = item.select("span.s-item__shipping.s-item__logisticsCost")
                    item_shipping = []
                    for i in shipping:
                        item_shipping.append(i.text)
                else:
                    item_shipping = ["Free Shipping"]
    

                print("Page " + str(g+1) + " Item " + str(j+1) + ": ", gf_title, item_price, item_shipping)
                
        
        
            
    except Exception as ex:
        print('error: ' + str(ex))


if __name__ == '__main__':
    main()
    
```
